```python
import pandas as pd

# DataFrame de práctica
df = pd.DataFrame({
    "Nombre": ["Ana", "Luis", "María", "Pedro", "Lucía", "Carlos", "Julia", "Erick"],
    "Matemática": [14, 17, 8, 12, 19, 13, 10, 16],
    "Historia":   [16, 14, 6, 10, 20, 9, 13, 12],
    "Inglés":     [12, 18, 4, 13, 15, 7, 11, 14]
})

df.set_index("Nombre", inplace=True)
```

```python
df
```

```python
#Filtra los estudiantes que no sacaron 12 en ninguna materia.
# Creamos la condición: si sacó 12 en alguna materia
condicion = (df[["Matemática", "Historia", "Inglés"]].eq(12)).any(axis=1)

# Nueva columna que indica si NO sacó 12 en ninguna
df["estudiante_condicion_12"] = ~condicion

# Mostramos solo esa columna
df["estudiante_condicion_12"]
```

```python
condiciones = df[["Matemática", "Historia", "Inglés"]].eq(20).any(axis=1)
df["condiciones"] = condiciones
df[df["condiciones"]==True] 
```

```python
#  Ejercicio 2 – Estudiantes con todas las notas mayores o iguales a 14
# Muestra solo a los estudiantes que hayan sacado 14 o más en todas las materias.

condiciones_1 = df[["Matemática", "Historia", "Inglés"]].ge(14).any(axis=1)
df["14"] = condiciones_1
mayor =df[df["14"]==True]
mayor
```

```python
#✅ Ejercicio 3 – Estudiantes que tienen exactamente un 12 en una sola materia
#Muestra solo los estudiantes que tienen un solo 12, y en el resto no.

condiciones_1 = df[["Matemática", "Historia", "Inglés"]].eq(12).sum(axis=1)==1
df["15"] = condiciones_1
mayor =df[df["15"]==True]
mayor
```

```python
# Ejercicio 4 – Estudiantes que no tienen ninguna nota menor a 10
# Filtra a quienes no tienen notas desaprobatorias (es decir, todo ≥10).

condiciones_1 = df[["Matemática", "Historia", "Inglés"]].le(10).any(axis=1)
df["16"] = condiciones_1
mayor =df[df["16"]==True]
mayor
```

```python
# Ejercicio 5 – Estudiantes cuyo promedio es menor que 13 y ninguna nota es menor que 10
#Muestra estudiantes con promedio bajo pero ninguna nota desaprobatoria.


# 1. Calcula el promedio de las materias
df["Promedio"] = df[["Matemática", "Historia", "Inglés"]].mean(axis=1)

# 2. Condición: promedio menor que 13
condicion_promedio = df["Promedio"] < 13

# 3. Condición: ninguna nota menor que 10 (es decir, todas >= 10)
condicion_notas = df[["Matemática", "Historia", "Inglés"]].ge(10).all(axis=1)

# 4. Combinamos ambas condiciones
condiciones_1 = condicion_promedio & condicion_notas

# 5. Nueva columna opcional para marcar los que cumplen
df["Cumple_Ejercicio_5"] = condiciones_1

# 6. Filtramos los estudiantes que cumplen con ambas condiciones
mayor = df[condiciones_1]

# Mostramos el resultado
mayor


```

```python
# Filtra a los estudiantes que sacaron entre 12 y 16 (inclusive) en todas las materias.
trece = df[["Matemática", "Historia", "Inglés"]].eq(13).any(axis=1)
diez = df[["Matemática", "Historia", "Inglés"]].le(10).any(axis=1)
total = trece & diez
df["ejercicio"] = total
df[total]
```
